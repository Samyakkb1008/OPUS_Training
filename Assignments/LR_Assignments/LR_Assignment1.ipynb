```python
def remove_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[column] >= lower) & (data[column] <= upper)]

for col in df.select_dtypes(include=np.number).columns:
    df = remove_outliers(df, col)
```

```python
split = int(0.8 * len(X))

X_train = X[:split]
X_test = X[split:]

y_train = y[:split]
y_test = y[split:]
```

```python
def compute_cost(X, y, w, b):    
    n = len(y)    
    y_pred = np.dot(X, w) + b    
    cost = np.sum((y - y_pred) ** 2) / n    
    return cost
```

```python
def gradient_descent(X, y, w, b, lr, iterations):    
    n = len(y)
    cost_history = []    
    for i in range(iterations):
        y_pred = np.dot(X, w) + b
        
        dw = (-2/n) * np.dot(X.T, (y - y_pred))
        db = (-2/n) * np.sum(y - y_pred)
        
        w = w - lr * dw
        b = b - lr * db
        
        cost = compute_cost(X, y, w, b)
        cost_history.append(cost)
        
    return w, b, cost_history
```

```python
learning_rates = [0.1, 0.01, 0.001]

results = {}

for lr in learning_rates:
    
    w = np.zeros(n_features)
    b = 0
    
    w, b, cost_history = gradient_descent(X_train, y_train, w, b, lr, 500)
    
    results[lr] = cost_history
    
    plt.plot(cost_history, label=f"lr={lr}")

plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost vs Iterations")
plt.legend()
plt.show()
```

```python
def predict(X, w, b):
    return np.dot(X, w) + b
```

```python
y_pred = predict(X_test, w, b)
MAE = np.mean(np.abs(y_test - y_pred))
MSE = np.mean((y_test - y_pred) ** 2)
RMSE = np.sqrt(MSE)
R2 = 1 - (np.sum((y_test - y_pred) ** 2) / np.sum((y_test - np.mean(y_test)) ** 2))

print("MAE:", MAE)
print("MSE:", MSE)
print("RMSE:", RMSE)
print("R2 Score:", R2)
```

```python
def stochastic_gradient_descent(X, y, lr=0.00001, epochs=50):

    n, m = X.shape
    
    w = np.zeros(m)
    b = 0
    
    for epoch in range(epochs):
        
        for i in range(n):
            
            xi = X[i]
            yi = y[i]
            
            y_pred = np.dot(xi, w) + b
            
            error = yi - y_pred
            
            dw = -2 * xi * error
            db = -2 * error
            
            w = w - lr * dw
            b = b - lr * db
            
    return w, b
```

```python
w_sgd, b_sgd = stochastic_gradient_descent(X_train, y_train)

y_pred_sgd = np.dot(X_test, w_sgd) + b_sgd

mse_sgd = np.mean((y_test - y_pred_sgd)**2)

print("SGD MSE:", mse_sgd)
```
